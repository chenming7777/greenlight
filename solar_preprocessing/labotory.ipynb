```python
import pandas as pd

# Read the CSV file
df = solar_weather = pd.read_csv("solar_weather_processed.csv")

# Remove Energy_delta_Wh column
df = df.drop('Energy delta_Wh', axis=1)

# Filter for June 1-7, 2018 (days 152-158)
filtered_df = df[
    (df['Year'] == 2018) & 
    (df['Month_num'] == 6) & 
    (df['DayOfYear'].isin([152, 153, 154, 155, 156, 157, 158]))
]

# Save the modified data to a new CSV file
filtered_df.to_csv("laboratory.csv", index=False)

# Display the first few rows to verify
filtered_df.head()
```
